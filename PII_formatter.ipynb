```python
MAIN_OUTPUT_PATH="/content/drive/MyDrive/nairaland_webtext/processed_formatted"
if not os.path.exists(MAIN_OUTPUT_PATH):
    os.makedirs(MAIN_OUTPUT_PATH)
```

```python
pipeline=[ JsonlReader(
            f"/content/drive/MyDrive/nairaland_webtext/processed_final/00000.jsonl.gz"
            ) ,
        PIIFormatter(),
        TokensCounter(),
        JsonlWriter(f"{MAIN_OUTPUT_PATH}"),
        #Jr(f"{MAIN_OUTPUT_PATH}"),
    ]
```
