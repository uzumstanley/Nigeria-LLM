```python
import pickle

def load_pickle_file(file_path):
    with open(file_path, 'rb') as file:  # Open the file in binary read mode
        data = pickle.load(file)
    return data

# This will display the contents of the pickle file
```

```python
def apply_trafilatura(text):
    return trafilatura.extract(text, favor_precision=True)
```

```python
files=[f'results{i}.pkl' for i in range(0,100)]
```

```python
results=[]
for file in files[0:33]:
    if file.endswith(".pkl"):
      print(file)
      list_data=load_pickle_file(f"/content/drive/MyDrive/nairaland_webtext/downloaded_pages/{file}")
      results.extend(list_data)
```

```python
data_df=pd.DataFrame(results)
```

```python
del list_data
del results
```

```python
data_403=data_df[data_df["error"]==403]#.shape
data_200=data_df[data_df["error"]==200]
```

```python
del data_df
```

```python
data_200.drop_duplicates("text",inplace=True)
```

```python
text_trafilatura=[]
for text in tqdm(list(data_200["text"])):
    #print(i)
    text_trafilatura.append(apply_trafilatura(text))
data_200["text"]=text_trafilatura
```
