```python
import gcld3
```

```python
detector=gcld3.NNetLanguageIdentifier(min_num_bytes=0, max_num_bytes=1000)

def detect_language_cld3(text):
    try:
        result = detector.FindLanguage(text=text)
        return result.language
    except Exception as e:
        print(text)
        return "error"

def detect_language_probability(text):
    try:
        result = detector.FindLanguage(text=text)
        return result.probability
    except Exception as e:
        print(text)
        return "error"
```

```python
dataset=dataset['train'].to_pandas()
```

```python
dataset["language"]=dataset["text"].apply(detect_language_cld3)
dataset["language_probability"]=dataset["text"].apply(detect_language_probability)
```

```python
dataset = Dataset.from_pandas(dataset)
```

```python
huggingface_hub.login()
```

```python
dataset.push_to_hub("naijaweb",private =True)
```
